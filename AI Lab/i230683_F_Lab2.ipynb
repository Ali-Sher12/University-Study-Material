```python
#question 1
import pandas as pd

data_ = pd.read_csv('students_scores_data.csv',na_values=' ')

mean1 = data_['Math'].mean(skipna=True)
mean2 = data_['Science'].mean(skipna=True)
mean3 = data_['English'].mean(skipna=True)
print(data_)
            
data_.dropna(thresh=3,inplace=True)
print(data_)

data_['Math'].fillna(mean1,inplace=True)
data_['Science'].fillna(mean2,inplace=True)
data_['English'].fillna(mean3,inplace=True)

print(data_)

for i in range(1,data_.shape[0]): #rows
    for j in range(2,5): #cols
        if data_.iloc[i,j] <= 0:
            data_.iloc[i,j] = 0
    
print(data_)

data_['Subjects\' mean'] = data_[['Math','Science','English']].mean(axis=1)
print(data_)

data_['Status'] = 1
data_.loc[data_['Subjects\' mean'] > 75, 'Status'] = 0
print(data_)

passed = data_['Status'].sum()
print('Total students that passed : ',passed)
print('Total students that passed %age : ',passed/(data_.shape[0]-1)*100)

data_.sort_values(by='Subjects\' mean',ascending=False).head(3)



```

```python
#question 2
map_of_the_color = []

for i in data_['English']:
    if i >= 90:
        map_of_the_color.append('green')
    elif i >= 75:
        map_of_the_color.append('orange')
    else:
        map_of_the_color.append('red')

data_.plot.scatter(x='Math',y='Science',s=data_['English'],c=map_of_the_color)
```
