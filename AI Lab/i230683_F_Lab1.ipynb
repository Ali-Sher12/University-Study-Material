```python
#Question 2

if __name__ == "__main__":

    #part a
    data_points = [(2, 5), (1, 3), (4, 8), (3, 6), (2, 7)]
    print("part a ",sorted(data_points,reverse=True, key=lambda arr: (arr[0]+arr[1])))
    
    #part b
    def isPrime(x):
        for i in range(2 , x):
            if i != x and x % i == 0:
                return False
        return True
        
    for i in data_points:
        if(isPrime(i[0]) and isPrime(i[1])):
            data_points.remove(i)
    print("part b ",data_points)
    
    #part c
    data_points = [(2, 5), (1, 3), (4, 8), (3, 6), (2, 7)]
    tempTouple = []
    for i in data_points:
        tempTouple.append(i[1])
    tempTouple = tuple(tempTouple)
    print("part c ",tempTouple)

    #part d
    product = 1
    for i in tempTouple:
        product = product * i
    print("part d ",product)    
    
```

```python
#Question 3


class Product:
    def __init__(self,prod,nam,price):
        self.product_id = prod
        self.name = nam
        self.price = price
    def display_details(self):
        print("product ID ",self.product_id)
        print("name ",self.name)
        print("price ",self.price)        

class Electronics(Product):
    def __init__(self,prod,nam,price,chip):
        self.chips = chip
        super().__init__(prod,nam,price)
    
    def display_details(self):
        super().display_details()
        print("No. of chips ",self.chips)

    def apply_discount(self,percentage):
        self.price = self.price - percentage * self.price    
    
class Clothing(Product):
    def __init__(self,prod,nam,price,fab):
        self.fabric = fab
        super().__init__(prod,nam,price)
    
    def display_details(self):
        super().display_details()
        print("Fabric Type ",self.fabric)    
    
    def apply_discount(self,percentage):
        self.price = self.price - percentage * self.price

if __name__ == "__main__":

    a = Electronics(1,"Fridge",1000,12)
    b = Electronics(2,"TV",1050,14)
    c = Clothing(3,"Zara",2000,"Polyester")
    d = Clothing(4,"Sana Safinaz",500,"Cotton")
    Kart = []
    Kart.append(a)
    Kart.append(b)
    Kart.append(c)
    Kart.append(d)
    myKart = []
    ch = 9
    while ch != -1:
        for i in Kart:
            print("------------------------")
            print(i.display_details())
            print("------------------------")
        print("Enter -1 to exit")
        ch = int(input("Enter Product ID "))
        if ch == -1:
            break
        Kart[ch].apply_discount(0.3)
        myKart.append(Kart[ch])
    print("My Kart :")
    for i in myKart:
        i.display_details()
        print("_________________________")
        

    
```

```python
#Question 4

class Book:
    def __init__(self,id,title_,av):
        self.book_id = id
        self.name = title_
        self.__availablity = av
        self.__borrowersList = []

    def checkAvailability(self):
        return self.__availablity
    
    def returnBook(self):#Question 5
        self.__availablity = True
        
    def BorrowBook(self,roll_no):
        self.__availablity = False
        self.__borrowersList.append(roll_no)

    def display_details(self):
        print("ID :" , self.book_id)
        print("Title :" , self.name)        
        print("Availablity :" , self.__availablity)        
        print("Borrowing History (RollNumbers):")
        print(self.__borrowersList)

if __name__ == "__main__":

    a = Book(1,"Arch Linux Tutorial",True)
    b = Book(2,"Shahmir: The legend of the rainbow",True)
    c = Book(3,"Help! My roommate is Mustafa Saleem",True)
    d = Book(4,"Freakanomics",True)
    Library = []
    Library.append(a)
    Library.append(b)
    Library.append(c)
    Library.append(d)

    ch = 9
    while ch != -1:
        for i in Library:
            print("------------------------")
            print(i.display_details())
            print("------------------------")
        print("Type -1 to exit")
        ch = int(input("Enter Book ID to borrow"))
        if Library[ch].checkAvailability() == False:
            print("Not Available")
        elif ch == -1:
            break
        else:
            nam = input("Enter Your roll no. ")        
            Library[ch].BorrowBook(nam)
            
    ch = 9
    while ch != -1:
        for i in Library:
            print("------------------------")
            print(i.display_details())
            print("------------------------")
        print("Type -1 to exit")
        ch = int(input("Enter Book ID to return"))
        if Library[ch].checkAvailability() == True:
            print("Already Present")
        elif ch == -1:
            break
        else:
            Library[ch].returnBook()        
```

```python
#Question 5


class Rectangle:
    def __init__(self,l,w):
        self.length = l
        self.width = w

    def Area(self):
        return self.length * self.width
    
    def Diagonal(self):
        return (self.length ** 2 + self.width ** 2) ** 0.5    
    
if __name__ == "__main__":
        
    rec = Rectangle(int(input("Length in feet ")),int(input("Width in feet ")))
    print("Area : ",rec.Area())
    print("Diagonal : ",rec.Diagonal())    

    
```

```python
#Question 6

class LibraryItem:
    def __init__(self,id,title_,av):
        self.item_id = id
        self.title = title_
        self.__availablity = av
        self.__borrowersList = []

    def checkAvailability(self):
        return self.__availablity
    
    def returnBook(self):
        self.__availablity = True
        
    def checkOut(self,roll_no):
        self.__availablity = False
        self.__borrowersList.append(roll_no)

    def display_details(self):
        print("ID :" , self.item_id)
        print("Title :" , self.title)        
        print("Availablity :" , self.__availablity)        
        print("Borrowing History (RollNumbers):")
        print(self.__borrowersList)
        
class Book(LibraryItem):
    def __init__(self,id,title_,av,auth):
        self.author = auth
        super().__init__(id,title_,av)
        
    def display_details(self):
        super().display_details()
        print("Author :" , self.author)        
        
    def calPenalty(self):
        return 100

class Thesis(LibraryItem):
    def __init__(self,id,title_,av,topic_):
        self.topic = topic_
        super().__init__(id,title_,av)        

    def display_details(self):
        super().display_details()
        print("Thesis topic :" , self.topic)                

if __name__ == "__main__":

    a = Book(1,"Arch Linux Tutorial",True,"Ali")
    b = Book(2,"Shahmir: The legend of the rainbow",True,"Abdullah")
    c = Thesis(3,"Thesis byMustafa Saleem",True,"Some topic")
    d = Thesis(4,"Thesis 2",True,"Random topic")

    Library = []
    Library.append(a)
    Library.append(b)
    Library.append(c)
    Library.append(d)

    ch = 9
    while ch != -1:
        for i in Library:
            print("------------------------")
            print(i.display_details())
            print("------------------------")
        print("Type -1 to exit")
        ch = int(input("Enter Book ID to borrow"))
        if Library[ch].checkAvailability() == False:
            print("Not Available")
        elif ch == -1:
            break
        else:
            nam = input("Enter Your roll no. ")        
            Library[ch].checkOut(nam)
            
    ch = 9
    while ch != -1:
        for i in Library:
            print("------------------------")
            print(i.display_details())
            print("------------------------")
        print("Type -1 to exit")
        ch = int(input("Enter Book ID to return"))
        if Library[ch].checkAvailability() == True:
            print("Already Present")
        elif ch == -1:
            break
        else:
            Library[ch].returnBook()        
```
